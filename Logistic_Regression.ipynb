```python
%matplotlib inline
import matplotlib as plt
import pandas as pd
import os
import numpy as np
```

```python
#import our train and test csv
train_path = os.path.join("Data/train.csv")
train_df = pd.read_csv(train_path)

test_path = os.path.join("Data/test.csv")
test_df = pd.read_csv(test_path)
```

```python
#convert the satisfaction column to 1's and 0's
train_df["satisfaction"].replace({"satisfied":1, "neutral or dissatisfied":0}, inplace=True)
```

```python
#assign the X_train data
X_train = np.array(train_df[['Flight Distance', 'Inflight wifi service',
       'Departure/Arrival time convenient', 'Ease of Online booking',
       'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
       'Inflight entertainment', 'On-board service', 'Leg room service',
       'Baggage handling', 'Checkin service', 'Inflight service',
       'Cleanliness', 'Departure Delay in Minutes', 'Arrival Delay in Minutes']]).reshape(-1,17)

#assign the y_train data
y_train = np.array(train_df["satisfaction"]).reshape(-1,1)
```

```python
X_train[0]
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
#convert the satisfaction column to 1's and 0's
test_df["satisfaction"].replace({"satisfied":1, "neutral or dissatisfied":0}, inplace=True)
```

```python
X_test = np.array(test_df[['Flight Distance', 'Inflight wifi service',
       'Departure/Arrival time convenient', 'Ease of Online booking',
       'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
       'Inflight entertainment', 'On-board service', 'Leg room service',
       'Baggage handling', 'Checkin service', 'Inflight service',
       'Cleanliness', 'Departure Delay in Minutes', 'Arrival Delay in Minutes']]).reshape(-1,17)
y_test = np.array(test_df["satisfaction"]).reshape(-1,1)
```

```python
y_test
```

```python
print(f'Training Data Score: {classifier.score(X_train, y_train) * 100}%') # 66%
print(f'Testing Data Score: {classifier.score(X_test, y_test) * 100}%') # 66%
```

```python
#predict the new data point
predictions = classifier.predict(X_test)
y_test_ = np.array(test_df["satisfaction"])

prediction_df = pd.DataFrame({"Prediction":predictions, "Actual":y_test_})
prediction_df
```
